```python
import pandas as pd
```

```python
#Đọc dữ liệu từ các tệp CSV
candidate_df = pd.read_csv("D:/Nhom_5_Case_6/Data_raw/candidate_profile.csv")
interview_df = pd.read_csv("D:/Nhom_5_Case_6/Data_raw/interview_score.csv")
salary_df = pd.read_csv("D:/Nhom_5_Case_6/Data_raw/salary_offer.csv")
```

```python
#Làm sạch candidate_id
candidate_df["candidate_id"] = (
    candidate_df["candidate_id"].astype(str).str.strip().str.upper())

#Làm sạch full_name
candidate_df["full_name"] = (
    candidate_df["full_name"].astype(str).str.strip().str.title())
```

```python
# Xử lý experience_years bị âm hoặc sai định dạng
candidate_df["experience_years"] = pd.to_numeric(
    candidate_df["experience_years"], errors='coerce')

#Giá trị âm thành 0
candidate_df["experience_years"] = candidate_df["experience_years"].where(candidate_df["experience_years"] >= 0, 0)

#Giá trị NaN thành 0
candidate_df["experience_years"] = candidate_df["experience_years"].where(candidate_df["experience_years"].notna(), 0)
```

```python
# Chuẩn hoá round và score
interview_df["round"] = pd.to_numeric(
    interview_df["round"],
    errors="coerce"
)

interview_df = interview_df[interview_df["round"].isin([1, 2, 3])]

interview_df["score"] = pd.to_numeric(
    interview_df["score"],
    errors="coerce"
)

interview_df = interview_df[
    (interview_df["score"] >= 0) & (interview_df["score"] <= 10)
]
```

```python
#Làm sạch offer_salary
salary_df["offer_salary"] = (
    salary_df["offer_salary"]
    .astype(str)
    .str.replace(",", "")
    .str.replace("VND", "")
    .str.replace("vnđ", "")
    .str.replace("đ", "")
    .str.strip()
)

salary_df["offer_salary"] = pd.to_numeric(
    salary_df["offer_salary"],
    errors="coerce"
)
```
